## Diving into Gemini API with Python 
### Generating Text From Images

```python
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv(), override=True) 
```

```python
import google.generativeai as genai
import os
genai.configure(api_key=os.environ.get('GOOGLE_API_KEY'))
```

```python
from PIL import Image
img = Image.open('files/benji.jpg')
img
```

```python
model = genai.GenerativeModel('gemini-2.5-flash')

text_prompt = 'Describe this image in detail.'
response = model.generate_content([text_prompt, img])
print(response.text)
```

```python
text_prompt = 'What is the brand of the black speakers?'
response = model.generate_content([text_prompt, img])
print(response.text)
```
