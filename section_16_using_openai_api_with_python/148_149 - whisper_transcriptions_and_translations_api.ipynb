## Transcriptions with Whisper

```python
# loading the OpenAI API key for authentication
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv(), override=True)
```

```python
from openai import OpenAI
client = OpenAI()

file = 'rr.mp3'
with open(file, 'rb') as audio_file: 
    transcript = client.audio.transcriptions.create(  # Send the file to OpenAI's Whisper model for transcription
        model='whisper-1',  # Specify the Whisper model for speech-to-text processing
        file=audio_file     # Pass the opened audio file as input
    )    
print(transcript)
```

```python
print(transcript.text)
```

## Translations with Whisper

```python
from openai import OpenAI
client = OpenAI()

with open('german.mp3', 'rb') as audio_file:
    translation = client.audio.translations.create(
        model='whisper-1',
        file=audio_file
    )

print(translation)
```

```python
print(translation.text)
```
