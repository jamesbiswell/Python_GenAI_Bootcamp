###  Chat Completions Endpoint

```python
from openai import OpenAI

# api_key = ''
# client = OpenAI(api_key=api_key)  # passing the api key explicitly

client = OpenAI()   # Initialize the OpenAI client, automatically loading the API key from the .env file

# Define the prompt that will be sent to the model
# prompt = 'Write a poem about a lonely robot exploring a futuristic city.'
prompt = 'Write a haiku about a robot dreaming of flowers.' 

response = client.chat.completions.create(
    model='gpt-4o-mini',  # Specify the OpenAI model to use for text generation
    messages=[            # Define the conversation history as a list of message dictionaries
        {'role': 'developer', 'content': 'You are a Sci-Fi writer and poet.'},  # Developer message guiding the modelâ€™s behavior ***
        {'role': 'user', 'content': prompt}  # User message containing the input prompt
    ], 
    temperature=1  # Set the randomness of the response (higher values make output more creative)
)

# *** In the latest API version, the system message was renamed to the developer message.
```

```python
print(response.choices[0].message.content)
```

```python
# listing available models
for item in client.models.list():
    print(item)
```

### Streaming API Responses

```python
from openai import OpenAI

client = OpenAI()

stream = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {'role': 'system', 'content': 'You high-school tutor.'},
        {'role': 'user', 'content': 'Explain Quantum Mechanics to teenagers in a short and concise way.'}
    ],
    stream=True # Enable streaming, so the response is returned in chunks instead of all at once
)

# Processing the streaming response
print('AI is typing...\n')
for chunk in stream:
    if chunk.choices[0].delta.content is not None:
        print(chunk.choices[0].delta.content, end='')
```
